# Estudio de la media en Python
###### recordemos que este codigo tambien lo podemos encontrar en su version en R 

```python
import random
import numpy as np
import matplotlib.pyplot as plot
f=[ ]
for i in range(50):
    b=f.append(random.randint( 1, 6))
b=np.array(f)
print(b)
len(b)
```

**Acabamos de generar un vector de 50 numeros aleatorios enteros del 1 al 6, lo cual corresponde a una simulacion de lanzar 50 veces un dato**

### Entonces generaremos el histograma de lanzar 50 dados


```python
plot.hist(x=b, color='lightgreen', rwidth=0.85)
plot.title('Histograma lanzar 50 veces un dado' )
plot.xlabel('Resultado')
plot.ylabel('Frecuencia')


plot.show() #dibujamos el histograma generado con matplotlib
```

#### Como podemos ver, solo son 50 veces, intentemos agrandar este numero y hacerlo mas veces a ver si tiende a algo

```python

f=[ ]
for i in range(50000):
    b=f.append(random.randint( 1, 6))
b=np.array(f)
len(b)
plot.hist(x=b, color='violet', rwidth=0.85)
plot.title('Histograma lanzar 50000 veces un dado')
plot.xlabel('Resultado')
plot.ylabel('Frecuencia')


plot.show() #dibujamos el histograma generado con matplotlib
```

### Podemos observar que todos los casos posibles tienden a tener 1/6 de las veces que se lanza un dado, precisamente porque es la probabilidad de que caiga cada uno

Ahora analizaremos el comportamiento de las medias, conforme el numero de veces que se lanza el dado va aumentando. 

```python
import random 
import numpy as np
f=[ ]
media= [ ]
mediaacum=[ ]
for i in range(1000):
    for i in range(50):
        b=f.append(random.randint( 1, 6))
    b=np.array(f)
    media=b.mean()
    mediaacum=np.append(media,mediaacum)
    medias=np.append(media,mediaacum)
    r=np.delete(medias,1)
len(r)
min(r)
```

```python
plot.hist(x=r, color='red', rwidth=0.85,bins=10)
plot.title('Histograma lanzar 50 veces un dado, calcular su media \n experimento repetido 1000 veces')
plot.xlabel('Resultado')
plot.ylabel('Frecuencia')


plot.show() #dibujamos el histograma generado con matplotlib
```
